```python
import pandas as pd
```

```python
data = pd.read_csv("../input/addition/add.csv")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['x'],data['sum'])
plt.xlabel("Independent")
plt.ylabel("sum(dependent)")
```

```python
plt.scatter(data['y'],data['sum'])
plt.xlabel("Independent")
plt.ylabel("sum(dependent)")
```

Here seeing the linear relationship between the 2 variables we are going to use the Linear regression.


# 1. Store the Feature Matrix in X and Response(Target) in Vector y

```python
X  = data[['x','y']]
Y = data['sum']
```

# 2. Train/Test Split
1. Split data into two part : a training set and a testing set
2. Train the model on training set
3. Test the Model on testing set 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

# 3. Import & Train The Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

# 4 Check Model's Prediction Performance

```python
model.score(X_train, Y_train)
```

# 5. Comparing The Results

```python
Y_pred = model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual':Y_test,'Prediction':Y_pred})
df
```

# 6. Prediction

```python
model.predict([[10,20]])
```

```python
model.predict([[100.2,210.3]])
```

# 7. Save Model Usign Joblib

Joblib is such an pacakge that can simply turn our Python code into parallel computing mode and of course increase the computing speed. Joblib is optimized to be fast and robust in particular on large data and has specific optimizations for numpy arrays.

```python
import joblib
```

```python
joblib.dump(model,'model_joblib')
```
